```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
female_data = pd.read_csv('ANSUR II FEMALE Public.csv',encoding='')
female_data.head()
```

```python
male_data = pd.read_csv('ANSUR II MALE Public.csv')
male_data.head()
```

```python
data = pd.concat([female_data,male_data])
data
```

```python
#preprocessing
data["weightkg"] = data["weightkg"] / 10  # convert to kg
data["stature"] = data["stature"] / 10    # convert to cm

data["Gender"] = data["Gender"].replace(["Female","Male"],[0,1])
 
```

```python
cmap = ["red","blue"]
plt.scatter(data["stature"],data["weightkg"],c=data["Gender"],alpha=0.1)
plt.show()
```
