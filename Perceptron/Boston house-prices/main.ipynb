Importing the required libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from my_pre import Perceptron
```

Reading the dataset and preprocessing it:

```python
data = pd.read_csv("Boston.csv")
data.head()
```

```python
data["area"] = data["crim"] * 100 + data["tax"]
data["price"] = data["area"] * 2 + data["age"] * -0.5 + 100 
```

Assigning values to X and Y:

```python
X = data[["age","area"]].values
Y = data["price"].values
Y = Y.reshape(-1,1)
```

Splitting the data for testing:

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

Starting the processing with the Perceptron model, displaying it on plt, and showing the loss:

```python
model = Perceptron(1,0.000001,60)
model.fit(X_train,Y_train)
y_pred = model.predict(X_test)
```

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_test[:, 1], X_test[:, 0], Y_test, color="green", label="Data")

ax.plot_trisurf(X_test[:, 1], X_test[:, 0], y_pred[:, 0], color="blue", alpha=0.5, label="Regression Line")

ax.set_xlabel("Area")
ax.set_ylabel("Age")
ax.set_zlabel("Price")
plt.title("3D Scatter Plot with Regression Line Perceptron")
plt.legend()

plt.show()
```

```python
loss = model.evaluate(X_test,Y_test)
print(loss)
```
