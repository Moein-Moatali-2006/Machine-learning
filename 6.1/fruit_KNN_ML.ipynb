import libraries :

```python
import numpy as np
import matplotlib.pyplot as plt
```

Make a Dataset for fruits:

```python
count = 100
# Apples
apple_width = np.random.normal(9,2,count)
apple_length = np.random.normal(9,2,count)

# Bananas
banana_width = np.random.normal(20,5,count)
banana_length = np.random.normal(6,1,count)

#Watermelons
watermelon_width = np.random.normal(30,5,count)
watermelon_length = np.random.normal(30,5,count)

# new fruit:
fruit_width_length = np.array([12,13])
```

show with scatter :

```python
plt.scatter(apple_width,apple_length,color="red")
plt.scatter(banana_width,banana_length,color="yellow")
plt.scatter(watermelon_width,watermelon_length,color="green")
plt.scatter(fruit_width_length[0],fruit_width_length[1],color="black")
plt.title("Diagram showing fruits")
plt.xlabel("width")
plt.ylabel("length")
plt.show()
```
