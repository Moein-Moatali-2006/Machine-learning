import libraries :

```python
import numpy as np
import matplotlib.pyplot as plt
```

Make a Dataset for fruits:

```python
count = 100
# Apples
apple_width = np.random.normal(9,2,count)
apple_length = np.random.normal(9,2,count)

# Bananas
banana_width = np.random.normal(15,5,count)
banana_length = np.random.normal(5,1,count)

#Watermelons
watermelon_width = np.random.normal(30,5,count)
watermelon_length = np.random.normal(30,5,count)

# new fruit:
fruit_width_length = np.array([8,14])
```

show with scatter :

```python
plt.scatter(apple_length,apple_width,color="red")
plt.scatter(banana_length,banana_width,color="yellow")
plt.scatter(watermelon_length,watermelon_width,color="green")
plt.scatter(fruit_width_length[0],fruit_width_length[1],color="black")
plt.title("Diagram showing fruits")
plt.xlabel("width(cm)")
plt.ylabel("length(cm)")
plt.legend(["Apple","Banana","Watermelon","unknown"])
plt.show()
```

start KNN :

```python
width = np.concatenate([apple_width,banana_width,watermelon_width])
length = np.concatenate([apple_length,banana_length,watermelon_length])

#A two-dimensional list of data:
X_train = np.column_stack((width,length))
```

```python
Y_apples = np.zeros(count,dtype="int")
Y_bananas = np.zeros(count,dtype="int") +1
Y_watermelons = np.zeros(count,dtype="int") +2

Y_train = np.concatenate([Y_apples,Y_bananas,Y_watermelons])
Y_train
```

### Euclidean Distance !

!["Euclidean_Distance"](Euclidean_Distance.png)

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))
```

Implementation of data :

```python
K = 5
```

```python
distances = []
for x in X_train:
    euclidean_number = euclidean_distance(fruit_width_length,x)
    distances.append(euclidean_number)
```

Fruit type index :

```python
nearest_neighbors = np.argsort(distances)[0:K]
nearest_neighbors
```

```python
Y_train[nearest_neighbors]
```

```python
result = np.bincount(Y_train[nearest_neighbors])
result
```

### show result :

```python
if np.argmax(result) == 0:
    guess_ML = "new_Apple"
    print("ðŸŽ")
elif np.argmax(result) == 1:
    guess_ML = "new_Banana"
    print("ðŸŒ")
elif np.argmax(result) == 2:
    guess_ML = "new_Watermelon"
    print("ðŸ‰")
```
