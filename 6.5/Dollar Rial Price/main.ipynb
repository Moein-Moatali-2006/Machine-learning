import libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lls import LinearLeastSquare
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

read and show data :

```python
data = pd.read_csv("Dollar_Rial_Price_Dataset.csv")
data.head()
```

Calculate of Ahmadinejad:

!["image](image\Mahmoud_Ahmadinejad.png)

```python
data["Date"] = pd.to_datetime(data["Date"])
# 2011 ---> 2013
df_Mahmod = data[(data["Date"].dt.year >= 2011) & (data["Date"].dt.year <= 2013)]
df_Mahmod["Close"] = pd.to_numeric(df_Mahmod["Close"].str.replace(",", "."))
max_Mahmod = df_Mahmod.sort_values(by="Close",ascending=False).head(1)
min_Mahmod = df_Mahmod.sort_values(by="Close",ascending=True).head(1)
```

```python
plt.figure(figsize=(12,6))
plt.title("Disaster of Ahmadinejad")
plt.plot(df_Mahmod["Date"],df_Mahmod["Close"],color="red")
plt.legend(["Dollar price (R)"])
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python

data['Date_numeric'] = (data['Date'] - data['Date'].min()).dt.days
data['Close'] = data['Close'].str.replace(',', '').astype(float)

X = data[['Date_numeric']].values  
Y = data['Close'].values 
```

```python
X_train , X_test , Y_train ,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
sk_lls = LinearRegression()
sk_lls.fit(X_train,Y_train)
Y_pred = sk_lls.predict(X_test)
```

Evaluate each model on test dataset using MAE loss function in Scikit-Learn:

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(Y_test,Y_pred)
```
