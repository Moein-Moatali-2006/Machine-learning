# Students Performance (Regression) LLS

Import libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from LLS import LLS
from scipy import stats
from sklearn.model_selection import train_test_split
```

generate dataset :

```python
np.random.seed(42)

hours = np.random.uniform(0,5,500)
grades = 5 * hours + np.random.normal(0,2,500)

grades = np.clip(grades, 0, 20)

data = pd.DataFrame({"Hour":hours , "Grade":grades})
```

split data:

```python
X = data["Hour"]
Y = data["Grade"]
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
```

show data and show some wrong line :

```python
plt.scatter(X_train,Y_train,color="green")
plt.plot(X_train,3*X_train,color="red")
plt.plot(X_train,2.5*X_train,color="red")
plt.plot(X_train,4*X_train,color="red")
plt.title("Displaying the Scores and Study Hours of Students")
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.show()
```

start LLS :

```python
lls = LLS()
lls.fit(X_train,Y_train)
lls.predict(X_test)
```

X * W = Y

```python
w = lls.w[1]
```

show result : 

```python
plt.figure(figsize=(12,6))
plt.scatter(X_test,Y_test,color="green")
plt.plot(X_test,w*X_test,color="blue")
plt.title("Displaying the Scores and Study Hours of Students on test data")
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.show()
```

using linregress from scipy library. 


```python
slope, intercept, r, p, std_err = stats.linregress(X_train, Y_train)
def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, X_test))

plt.scatter(X_test, Y_test)
plt.plot(X_test, mymodel,color="green")
plt.title("using linregress from scipy library")
plt.xlabel("study hours")
plt.ylabel("Scores")
plt.show()
```
